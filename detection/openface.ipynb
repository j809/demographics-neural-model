```python
import openface
import cv2

%pylab inline
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 256
import matplotlib.pyplot as plt
import matplotlib.image as mpimg


img = cv2.imread('picture.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.subplot(1,3,1)
plt.imshow(img)
plt.axis('off')

alignment = openface.AlignDlib('landmarks.dat')
face = alignment.getLargestFaceBoundingBox(img)
x = face.left()
y = face.top()
w = face.width()
h = face.height()

cv2.rectangle(img, (x,y), (x+w,y+h), (255, 0, 0), 1)

plt.subplot(1,3,2)
plt.imshow(img)
plt.axis('off')

plt.subplot(1,3,3)
crop_img = img[y:y+h, x:x+w]
plt.imshow(crop_img)
plt.axis('off')
```

```python
mpl.rcParams['figure.dpi'] = 256
group_img = cv2.imread('group.jpg')
group_img = cv2.cvtColor(group_img, cv2.COLOR_BGR2RGB)

plt.subplot(2,1,1)
plt.imshow(group_img)
plt.axis('off')

alignment = openface.AlignDlib('landmarks.dat')
faces = alignment.getAllFaceBoundingBoxes(group_img)

for face in faces:    
    x = face.left()
    y = face.top()
    w = face.width()
    h = face.height()

    cv2.rectangle(group_img, (x,y), (x+w,y+h), (255, 0, 0), 2)

plt.subplot(2,1,2)
plt.imshow(group_img)
plt.axis('off')
```

```python
mpl.rcParams['figure.dpi'] = 256
for index, face in enumerate(faces):
    x = face.left()
    y = face.top()
    w = face.width()
    h = face.height()
    
    plt.subplot(1, len(faces), index+1)
    crop_img = group_img[y:y+h, x:x+w]
    plt.imshow(crop_img)
    plt.axis('off')
```
